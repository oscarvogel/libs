## Parte 3 - Visualizacion de casos

En esta parte del tutorial puede encontrar algunas de las funciones de la clase *Cases Plot*, en particular esta clase hereda las propiedades de la clase Matplotlib.

Como de costumbre, el primer paso será importar el módulo arcovid19.

```python
import arcovid19 as ac
import matplotlib.pyplot as plt
```

La siguiente celda realiza una limpieza de caché, ya que arcovid19 guarda un chache de una hora por cada solicitud realizada en la base de datos en línea. Con esta acción, nos aseguramos de que estamos utilizando una base de datos actualizada.

```python
ac.CACHE.clear()
```

Cargamos la base de datos COVID-19 de Argentina.

```python
cases = ac.load_cases()
```

El método *plot* muestra la tendencia hasta la fecha de cada una de las provincias junto con la de Argentina. Por defecto, muestra casos confirmados, pero también puede elegir ver casos activos, recuperados y muertos.

```python
fig, ax = plt.subplots(figsize=(20, 8))
cases.plot(ax=ax, confirmed=False, active=True, recovered=False, deceased=False)
ax.legend(loc=2)
```

Luego, si está interesado en una provincia en particular, solo tiene que agregar *grate_full_periode* y especificar el nombre de la provincia. Si no ingresa el nombre de ninguna provincia, la herramienta le mostrará los casos de Argentina.

```python
ax = cases.plot.grate_full_period('Bs As', confirmed=True, active=True, recovered=True, deceased=True)
```

Además, si desea observar el comportamiento de las observaciones por fecha, puede usar las siguientes representaciones gráficas.

```python
cases.plot.time_serie_all()
```

```python
cases.plot.time_serie()
```

```python
cases.plot.barplot()
```

```python
cases.plot.boxplot(showfliers=False)
```
